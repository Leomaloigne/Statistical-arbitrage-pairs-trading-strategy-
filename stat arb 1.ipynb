```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
from itertools import combinations
from statsmodels.tsa.stattools import coint


def pair_analysis(t1, t2):

    df1 = yf.download(t1, start= '2017-01-01', end= '2025-01-01')

    df2 = yf.download(t2, start= '2017-01-01', end= '2025-01-01')

    data1 = df1['Close']
    data2 = df2['Close']

    prices = pd.concat([data1, data2], axis = 1).dropna()
    prices.columns = [t1, t2]

    score, p_value, _ = coint(prices[t1], prices[t2])

    return {
        'pair': (t1,t2),
         'p_value': p_value,
    }

test_tickers = ['SMT.L', 'ICG.L', 'RMV.L', 'EXPN.L', 'VOD.L', 'CPG.L', 'BATS.L', 'TSCO.L', 'ENT.L', 'CNA.L', 'CCH.L', 'SSE.L']
p_values = []

for t1, t2 in combinations(test_tickers, 2):
    value = pair_analysis(t1, t2)
    p_values.append(value)

p_values_df = pd.DataFrame(p_values).sort_values(by='p_value')

print(p_values_df.head())

significant_pairs = p_values_df[p_values_df['p_value']<0.05]

print('sig pairs:', len(significant_pairs))

print(significant_pairs)
```
